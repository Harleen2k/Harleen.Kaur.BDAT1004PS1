## Q6:
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the CSV file
read_data = pd.read_csv(r"C:\Users\harle\Downloads\us-marriages-divorces-1867-2014.csv")

# Plotting marriages and divorces data
plt.plot(read_data.Year, read_data['Marriages'], label='Marriages')
plt.plot(read_data.Year, read_data['Divorces'], label='Divorces')

# Adding title and labels to the chart
plt.title("Divorces and Marriages per capita in U.S. between 1867-2014")
plt.ylabel("Number of Marriages/Divorces")
plt.xlabel("Year")

# Adding a legend to the chart
plt.legend()

# Display the plot
plt.show()
```

## Q7:
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000. Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already read the data into the read_data DataFrame

# Filtering data for the years 1900, 1950, and 2000
graph = read_data[read_data['Year'].isin([1900, 1950, 2000])]

# Creating a bar plot
b_graph = graph.plot(x="Year", y=["Marriages", "Divorces"], kind="bar")

# Setting title, x-axis label, and y-axis label
b_graph.set_title('Number of Marriages and Divorces per capita in the U.S. between 1900, 1950 and 2000')
plt.xlabel('Year')
plt.ylabel('Number of Marriages & Divorces')

# Display the plot
plt.show()
```

## Q9:
Create a pie chart showing the fraction of all Roman Emperors that were assassinated. Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
#reading data from local disk
roman_data = pd.read_csv(r"C:\Users\harle\Downloads\roman-emperor-reigns.csv")
roman= pd.DataFrame({'count' : roman_data.groupby("Cause_of_Death" ).size()}).reset_index() 
labels = roman['Cause_of_Death'].values
sizes = roman['count'].values
x=0.0
y=0.1
explode=[y if item=='Assassinated' else x for item in labels ]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=0, textprops={'fontsize':12})
ax1.axis('equal')  
plt.show()
```
